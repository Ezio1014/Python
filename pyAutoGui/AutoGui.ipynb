```python
#執行時間
def time_now():
    localtime = time.localtime()
    result = time.strftime("%Y-%m-%d %I:%M:%S %p", localtime)
    print(result)

def click_off():
    time.sleep(0.1)
    pag.moveTo(1461,127)
    time.sleep(0.05)
    pag.click()

def self(times):    
    for t in range(times):
    #造船測試(OK)
        time.sleep(0.3)
        pag.moveTo(252,399)
        time.sleep(0.1)
        pag.click()
        time.sleep(0.1)
        pag.moveTo(466,308)
        time.sleep(0.1)
        pag.click()
        time.sleep(0.25)

        for i in range(30):
            time.sleep(0.15)
            pag.press("num1")
            time.sleep(0.08)
            pag.press("num2")
            time.sleep(0.08)
            pag.press("num2")    

        click_off()

    #賣船測試(OK)
        time.sleep(0.3)
        pag.moveTo(248,328)
        time.sleep(0.1)
        pag.click()
        time.sleep(0.1)
        pag.moveTo(469,241)
        time.sleep(0.1)
        pag.click()
        time.sleep(0.1)
        pag.moveTo(1517,372)
        time.sleep(0.1)
        pag.click()
        time.sleep(0.25)
        
        for i in range(31):
            time.sleep(0.2)
            pag.press("num4")
            time.sleep(0.15)
            pag.press("num5")
            time.sleep(0.15)
            pag.press("num6")

        click_off()
        
#火炮製作
def Artillery(times):
    for t in range(times):
        time.sleep(0.5)
        pag.moveTo(1528,346)
        time.sleep(0.3)
        pag.click()
        time.sleep(0.3)
        pag.moveTo(1462,332)
        time.sleep(0.3)
        pag.click()
        time.sleep(0.3)
        pag.moveTo(795,690)
        time.sleep(0.5)
        pyautogui.dragTo(795, 220, duration=1)
        time.sleep(1.5)
        pag.press("num7")        
        time.sleep(0.8)
        pag.press("num1")
        time.sleep(0.5)
        pag.press("num2")
        time.sleep(0.5)
        pag.press("num3")
        time.sleep(0.5)
        pag.press("num4")
        
        for i in range(20):
            time.sleep(0.5)
            pag.press("num5")
            time.sleep(0.5)
            pag.press("num6")
            time.sleep(0.5)
            pag.press("num7")
            time.sleep(1)
            pag.press("num7")
            time.sleep(0.3)
            pag.press("num5")
            time.sleep(0.3)
            pag.press("num5")
            time.sleep(0.3)
            pag.press("num5")
            time.sleep(0.3)
            pag.press("num5")
            time.sleep(0.3)
            pag.press("num5")
        
        time.sleep(1)
        pag.press("0")
        time.sleep(0.5)
        pag.press("0")
        time.sleep(0.8)
        pag.press("m")
        time.sleep(0.8)
        pag.moveTo(248,328)
        time.sleep(0.2)
        pag.click()
        time.sleep(0.2)
        pag.moveTo(465,298)
        time.sleep(0.2)
        pag.click()
        time.sleep(0.5)
        pag.press("num8")
        time.sleep(0.5)
        pag.press("num9")
        time.sleep(0.5)
        pag.press("num2")
        time.sleep(0.5)
        pag.press("num0")
        time.sleep(0.8)
        pag.press("0")
        time.sleep(0.5)
        pag.press("0")
```

```python
#num0  貿易商
#num1  交易
#num2  世界
#num3  導航
#空格  出港

#交易介面
def trading():
    pag.press("m")
    time.sleep(1.5)
    pag.press("num0")
    time.sleep(0.8)
    pag.press("num1")
    time.sleep(1)
    pag.press("num4")
    time.sleep(1)
    pag.press("num5")
    time.sleep(1)
    pag.press("num6")
    time.sleep(1)
    pag.press("num7")
#出港介面
def out_port():
    time.sleep(4)
    pag.press("num2")
    time.sleep(4)
#導航介面
def navigation():
    time.sleep(2)
    pag.press("num3")
    time.sleep(2)
    pag.press("space")
#縮小地圖
def zoomOutMap(t):
    time.sleep(3)
    for i in range(t):
        time.sleep(1.2)
        pag.press("num9")
    time.sleep(2)
#出售貨物
def selfGoods():
    pag.press("m")
    time.sleep(1.5)
    pag.press("num0")
    time.sleep(0.8)
    pag.press("num1")
    time.sleep(1)
    pag.press("space")
    time.sleep(1)
    pag.press("num8")
    time.sleep(1)
    pag.press("num7")
```
